#### RMSE 与 MAE

```python
from math import sqrt
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
def rmse(prediction,truth):
    prediction = prediction[truth.nonzero()].flatten()
    truth = truth[truth.nonzero()].flatten()
    return sqrt(mse(prediction,truth))
```

```python
def mae(prediction,truth):
    prediction = prediction[truth.nonzero()].flatten()
    truth = truth[truth.nonzero()].flatten()
    return mean_absolute_error(prediction,truth)
```

### 计算余弦相似度

```python
from sklearn.metrics.pairwise import pairwise_distances
```

```python
# 计算用户相似度
y_user_similarity = pairwise_distances(y_train_data_matrix, metric='cosine')
```

```python
# 计算餐厅相似度
y_item_similarity = pairwise_distances(y_train_data_matrix.T, metric='cosine')
```

```python
# 计算预测值
def predict(matrix, similarity, sim_type):
    # 基于用户相似度
    if sim_type == 'user':
        mean_user_ratings = matrix.mean(axis=1)
        # 减去每个用户评分平均值以消除不同用户不同评分习惯的影响
        matrix_diff = (matrix - mean_user_ratings[:,np.newaxis])
        # 预测时再加上平均分
        pred = mean_user_ratings[:,np.newaxis] + similarity.dot(matrix_diff)/np.array([np.abs(similarity).sum(axis=1)]).T
    # 基于物品相似度
    elif sim_type == 'item':
        pred = matrix.dot(similarity)/np.array([np.abs(similarity).sum(axis=1)])
    return pred
```

### 构造一个函数观察多方面子评分的预测效果

```python
def cf(feature):
    
    train_data_matrix = np.zeros((n_users,n_items))
    test_data_matrix = np.zeros((n_users,n_items))
    
    for line in train_data.itertuples():
        i = int(np.argwhere(users==line[1]))
        j = int(np.argwhere(items==line[2]))
        truth = line[feature]
        train_data_matrix[i,j] = truth
    for line in test_data.itertuples():
        i = int(np.argwhere(users==line[1]))
        j = int(np.argwhere(items==line[2]))
        truth = line[feature]
        test_data_matrix[i,j] = truth
    
    user_similarity = pairwise_distances(train_data_matrix, metric='cosine')
    item_similarity = pairwise_distances(train_data_matrix.T, metric='cosine')
    
    itembased_cf_prediction = predict(train_data_matrix,item_similarity,sim_type='item')  
    userbased_cf_prediction = predict(train_data_matrix,user_similarity,sim_type='user')
    
    userbasedCF_rmse = rmse(userbased_cf_prediction,test_data_matrix)
    itembasedCF_rmse = rmse(itembased_cf_prediction,test_data_matrix)
    print('对于%s评分'%train_data.iloc[:,feature-1].name+'的CF推荐，其RMSE是:')
    print('基于用户的CF的RMSE：'+str(userbasedCF_rmse)+'基于餐厅的CF的RMSE：'+str(itembasedCF_rmse))
    
    userbasedCF_mae = mae(userbased_cf_prediction,test_data_matrix)
    itembasedCF_mae = mae(itembased_cf_prediction,test_data_matrix)
    print('对于%s评分'%train_data.iloc[:,feature-1].name+'的CF推荐，其MAE是:')
    print('基于用户的CF的MAE：'+str(userbasedCF_mae)+'基于餐厅的CF的MAE：'+str(itembasedCF_mae))
    
    return train_data_matrix,test_data_matrix,userbasedCF_rmse,itembasedCF_rmse,userbasedCF_mae,itembasedCF_mae
```

# 实现PMF概率矩阵分解

```python
def SGD(train_data,test_matrix,N,M,eta,K,lambda_1,lambda_2,Step):
    # train: train data
    # test: test data
    # N:the number of user
    # M:the number of item
    # eta: the learning rata
    # K: the number of latent factor
    # lambda_1,lambda_2: regularization parameters
    # Step: the max iteration
    
    U = np.random.normal(0,0.1,(N,K))
    V = np.random.normal(0,0.1,(M,K))
    
    rmseList=[]
    maeList=[]
    loss=[]
    
    for ste in range(Step):
        los =0.0
        for line in train_data.itertuples():
            i = int(np.argwhere(users==line[1]))
            j = int(np.argwhere(items==line[2]))
            r = line[3]
            
            e = r-np.dot(U[i],V[j].T)
            U[i] = U[i] + eta*(e*V[j]-lambda_1*U[i])
            V[j] = V[j] + eta*(e*U[i]-lambda_2*V[j])

            los=los+0.5*(e**2+lambda_1*np.square(U[i]).sum()+lambda_2*np.square(V[j]).sum())
        
        loss.append(los)
        rms=rmse(np.dot(U,V.T),test_matrix)
        ma=mae(np.dot(U,V.T),test_matrix)
        rmseList.append(rms)
        maeList.append(ma)
#         if loss<L:
#             break
#         if ste%10==0:
#             print(ste//10)
    return loss,rmseList,maeList,U,V
    
```

```python

loss,rmseList,maeList,U,V=SGD(train_data,y_test_data_matrix,n_users,n_items,0.005,10,0.1,0.1,100)
```

## 使用NMF矩阵分解

```python
def nmf(V,test_matrix,k,r,e):
    m,n = np.shape(V)
    W=np.array(np.random.random((m,k)))
    H=np.array(np.random.random((k,n)))
    
    rmseList=[]
    maeList=[]
    
    for x in range(r):
        V_pred = np.dot(W,H)
        E=V-V_pred
        err = 0.0

        err = (E**2).sum()
        if err<e:
            break
        
        a=np.dot(W.T,V)
        b=np.dot(np.dot(W.T,W),H)
        
        for i_1 in range(k):
            for j_1 in range(n):
                if b[i_1,j_1] != 0:
                    H[i_1,j_1] = H[i_1,j_1]* a[i_1,j_1] / b[i_1,j_1]
        
        c=np.dot(V,H.T)
        d=np.dot(np.dot(W,H),H.T)
        for i_2 in range(m):
            for j_2 in range(k):
                if d[i_2, j_2] != 0:
                    W[i_2,j_2] = W[i_2,j_2] * c[i_2,j_2] / d[i_2, j_2]
        
        rms=rmse(np.dot(W,H),test_matrix)
        ma=mae(np.dot(W,H),test_matrix)
        rmseList.append(rms)
        maeList.append(ma)
        
    return W,H,rmseList,maeList
```

```python
def haversine(point1,point2):
    lat1,lon1,lat2,lon2 = point1[0],point1[1],point2[0],point2[1]
    # 将十进制度数转化为弧度
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1,lon2, lat2])
   
    dlon = lon2 - lon1  
    dlat = lat2 - lat1   
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2 
    c = 2 * asin(sqrt(a)) 
    r = 6371 # 地球平均半径，单位为公里  
    return round(c * r,2)  
```

```python
def bnmf(train_matrix,test_matrix,Step,K,sigma2):
    rmseList=[]
    maeList=[]
    
    I = n_users
    J = n_items
    
    U = np.random.exponential(alpha,(I,K))
    V = np.random.exponential(beta,(K,J))
    
    
    Y = train_matrix
    
    for ste in range(Step): 
        
        C = np.dot(V,V.T)
        D = np.dot(Y,V.T)
        
        for k in range(K):
            unk = np.concatenate((U[:,:k],U[:,k+1:]),axis=1)
            cnk = np.concatenate((C[:k,k],C[k+1:,k]),axis=0)
            an = (D[:,k] - np.dot(unk,cnk) -alpha*sigma2)/(C[k,k]+np.spacing(1))
            U[:,k] = pnn(an)
        
        chi = np.square(Y).sum()/2
        xi = (U*(np.dot(U,C)-2*D)).sum()/2
            
        sigma2 =(chi+xi)/(0.5*I*J+1)
        
        E = np.dot(U.T,U)
        F = np.dot(U.T,Y)
        
        for k in range(K):
            vnk = np.concatenate((V[:k,:],V[k+1:,:]),axis=0)
            enk = np.concatenate((E[k,:k],E[k,k+1:]))
            bn = (F[k,:] - np.dot(enk,vnk) -beta*sigma2)/(E[k,k]+np.spacing(1))
            V[k] = pnn(bn)
           
        rms=rmse(np.dot(U,V),test_matrix)
        ma=mae(np.dot(U,V),test_matrix)
        rmseList.append(rms)
        maeList.append(ma)
                
    return U,V,rmseList,maeList
```

```python
def b2nmf(train_matrix,test_matrix,Step,K,sigma2):
    rmseList=[]
    maeList=[]
    
    I = n_users
    J = n_items
    
    U = np.random.exponential(alpha,(I,K))
    V = np.random.exponential(beta,(K,J))
    
    Y = train_matrix
    dltepho = dlte_matrix*pho_matrix
    
    for ste in range(Step): 
        
        C= np.dot(V,V.T)
        D = np.dot(Y*dltepho,V.T)
        
        for k in range(K):
            unk = np.concatenate((U[:,:k],U[:,k+1:]),axis=1)
            vnk = np.concatenate((V[:k,:],V[k+1:,:]),axis=0)
            an = (D[:,k] - np.dot(np.dot(unk,vnk),V[k]) -alpha*sigma2)/(np.dot(dltepho,(V[k]**2).T)+np.spacing(1))
            U[:,k] = pnn(an)
        
        chi = np.square(Y).sum()*0.5
        xi = (U*(np.dot(np.dot(U,V)*dltepho,V.T)-2*np.dot(Y,V.T))).sum()/2
            
        sigma2 =(chi+xi)/(0.5*I*J+1)
        
        F = np.dot(U.T,Y*dltepho)
        
        for k in range(K):
            unk = np.concatenate((U[:,:k],U[:,k+1:]),axis=1)
            vnk = np.concatenate((V[:k,:],V[k+1:,:]),axis=0)
            bn = (F[k,:] - np.dot(U[:,k],np.dot(unk,vnk)) -beta*sigma2)/(np.dot(U[:,k]**2,dltepho)+np.spacing(1))
            V[k] = pnn(bn)
           
        rms=rmse(np.dot(U,V),test_matrix)
        ma=mae(np.dot(U,V),test_matrix)
        rmseList.append(rms)
        maeList.append(ma)
                
    return U,V,rmseList,maeList
```

```python
def gcbconmf(y_train_matrix,g_train_matrix,h_train_matrix,o_train_matrix,test_matrix,Step,K,sigma2):
    
    rmseList=[]
    maeList=[]
    
    I = n_users
    J = n_items
    
    U = np.random.exponential(alpha,(I,K))
    V = np.random.exponential(beta,(K,J))
    
    C = np.random.exponential(beta,(K,J))
    E = np.random.exponential(beta,(K,J))
    S = np.random.exponential(beta,(K,J))
    
    Y = y_train_matrix
    G = g_train_matrix
    H = h_train_matrix
    O = o_train_matrix
    
    dltepho = pho_matrix*dlte_matrix
    
    for ste in range(Step): 
        
        D = np.dot(Y*dltepho,V.T)
        CV = np.dot(V,V.T)
        CC = np.dot(C,C.T)
        CE = np.dot(E,E.T)
        CS = np.dot(S,S.T)
        
        DY = np.dot(Y,V.T)
        DH = np.dot(H,C.T)
        DG = np.dot(G,E.T)
        DO = np.dot(O,S.T)
        
        for k in range(K):
            unk = np.concatenate((U[:,:k],U[:,k+1:]),axis=1)
            vnk = np.concatenate((V[:k,:],V[k+1:,:]),axis=0)
            ccnk = np.concatenate((CC[:k,k],CC[k+1:,k]),axis=0)
            cenk = np.concatenate((CE[:k,k],CE[k+1:,k]),axis=0)
            csnk = np.concatenate((CS[:k,k],CS[k+1:,k]),axis=0)
            an = (D[:,k] - np.dot(np.dot(unk,vnk),V[k])+ DH[:,k] - np.dot(unk,ccnk)- DG[:,k] - 
                  np.dot(unk,cenk)+ DO[:,k] - np.dot(unk,csnk)-alpha*sigma2)/(np.dot(dltepho,(V[k]**2).T)+
                                                                             CC[k,k]+CE[k,k]+CS[k,k])
            U[:,k] = pnn(an)
        
        chi = np.square(Y).sum()*0.5
        xi = (U*(np.dot(np.dot(U,V)*dltepho,V.T)-2*np.dot(Y,V.T))).sum()/2
            
        sigma2 =(chi+xi)/(2*I*J+1)
        
        F = np.dot(U.T,Y*dltepho)
        
        for k in range(K):
            unk = np.concatenate((U[:,:k],U[:,k+1:]),axis=1)
            vnk = np.concatenate((V[:k,:],V[k+1:,:]),axis=0)
            bn = (F[k,:] - np.dot(U[:,k],np.dot(unk,vnk)) -beta*sigma2)/(np.dot(U[:,k]**2,dltepho)+np.spacing(1))
            V[k] = pnn(bn)
        
        
        Eu = np.dot(U.T,U)
        L = np.dot(U.T,H)
        
        for k in range(K):
            cnk = np.concatenate((C[:k,:],C[k+1:,:]),axis=0)
            eunk = np.concatenate((Eu[k,:k],Eu[k,k+1:]))
            cn = (L[k,:] - np.dot(eunk,cnk) -beta*sigma2)/(Eu[k,k]+np.spacing(1))
            C[k] = pnn(cn)
        
        M = np.dot(U.T,G)
        
        for k in range(K):
            enk = np.concatenate((E[:k,:],E[k+1:,:]),axis=0)
            eunk = np.concatenate((Eu[k,:k],Eu[k,k+1:]))
            en = (M[k,:] - np.dot(eunk,enk) -beta*sigma2)/(Eu[k,k]+np.spacing(1))
            E[k] = pnn(en)
        
        N = np.dot(U.T,O)
        
        for k in range(K):
            snk = np.concatenate((S[:k,:],S[k+1:,:]),axis=0)
            eunk = np.concatenate((Eu[k,:k],Eu[k,k+1:]))
            sn = (N[k,:] - np.dot(eunk,snk) -beta*sigma2)/(Eu[k,k]+np.spacing(1))
            S[k] = pnn(sn)
        
           
        rms=rmse(np.dot(U,V),test_matrix)
        ma=mae(np.dot(U,V),test_matrix)
        rmseList.append(rms)
        maeList.append(ma)
                
    return U,V,rmseList,maeList
```
